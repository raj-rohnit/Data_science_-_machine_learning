```python
def no_of_eps(text):
    pointer=0
    data=""
    for i in text:
        if i=="(":
            pointer=1
        if pointer==1:
            data=data+i
        if i==")":
            pointer=0
    numbers = ''.join(ch for ch in data if ch.isdigit())
    return numbers
```

```python
dataset["eps"]=dataset["Title"].apply(no_of_eps).astype(int)
dataset.head(5)
```

```python
dataset["Time_Interval"] = dataset["Title"].str.split(")").str[1].str[:19]
```

```python
# dataset["Title"].str.split("(").str[0].str[: dataset["Title"].str.find("(")-1]
dataset["Title"]=dataset["Title"].str.extract(r'^(.*?)\(')
```

```python
from datetime import datetime

def count_months(date_range):
    start_str, end_str = date_range.split(" - ")
    start_date = datetime.strptime(start_str, "%b %Y")
    end_date = datetime.strptime(end_str, "%b %Y")
    months = (end_date.year - start_date.year) * 12 + (end_date.month - start_date.month)
    
    return months
```

```python
dataset["no of months"]=dataset["Time_Interval"].apply(count_months)
```
