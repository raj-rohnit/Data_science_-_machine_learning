## dataframe 

```python
dataset = {
    "Name": ["Rahul", "Priya", "Amit", "Sneha", "Arjun"],
    "Age": [25, 30, 28, 32, 27],
    "City": ["Mumbai", "Delhi", "Bangalore", "Hyderabad", "Chennai"],
    "Salary": [50000, 65000, 72000, 80000, 54000]
}
pd.DataFrame(dataset,index=range(1,len(dataset["Name"])+1))
```

```python

dataset2 = {
    'name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank'],
    'age': [28, 34, 29, 31, 26, 35],
    'city': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Miami'],
    'salary': [75000, 85000, 72000, 90000, 68000, 80000]
}

pd.DataFrame(dataset2,index=range(1,len(dataset2["name"])+1))
```

```python
dataset3= [
    ['Alice', 28, 'New York', 75000],
    ['Bob', 34, 'Los Angeles', 85000],
    ['Charlie', 29, 'Chicago', 72000],
    ['Diana', 31, 'Houston', 90000],
    ['Eve', 26, 'Phoenix', 68000],
    ['Frank', 35, 'Miami', 80000]
]
df = pd.DataFrame(dataset3,index=range(1,len(dataset3)+1),columns=["Name", "Age", "City", "Salary"])
df
```

adding and deleting  a column


```python
df["kids"] = [0, 1, 2, 1, 0, 3]
```

```python
df.drop(["City", "kids"], axis=1, inplace=False)
# inplace=False means it will not change the original dataframe, it will return a new dataframe without the "City" column. If inplace=True, it will modify the original dataframe and remove the "City" column permanently.
# inplace=True will change the original dataframe permanently, so use it with caution. If you want to keep the original dataframe and create a new one without the "City" column, use inplace=False.
# axis=1 means we want to drop a column, if axis=0, it means we want to drop a row. In this case, we want to drop the "City" column, so we use axis=1. If we wanted to drop a row, we would use axis=0 and specify the index of the row we want to drop.
```

selecting a row

```python
df.loc[1:3, ["Name", "Salary"]]
# loc is used to access a group of rows and columns by labels or a boolean array. In this case, we are accessing rows 1 to 3 (inclusive) and the columns "Name" and "Salary". This will return a new dataframe with the selected rows and columns.
```

```python
df.loc[:,["Name", "City"]]
```

```python
df.iloc[0:3, 0:2]
# iloc is used to access a group of rows and columns by integer position.
#  In this case, we are accessing rows 0 to 2 (exclusive) and columns 0 to 1 (exclusive). 
# This will return a new dataframe with the selected rows and columns based on their integer position in the original dataframe.
```

conditional selection

```python
df[(df["Age"] > 30) & (df["Salary"] > 80000)]
```

finding the empty data 

```python
dataset4 = {
    "Name": ["Alice", None, "Charlie", "Diana", "Eve", None],
    "Age": [28, 34, None, 31, None, 35],
    "City": ["New York", "Los Angeles", "Chicago", None, "Phoenix", None],
    "Salary": [75000, None, 72000, 90000, None, 80000]}
df2=pd.DataFrame(dataset4,index=range(1,len(dataset4["Name"])+1))
```

```python
df2.isna()
```

```python
df2.isnull().sum()
```

```python
df2.isnull().any()
```

```python
df2.dropna( thresh=3,inplace=False)
#thresh=3 means that we want to drop rows that have less than 3 non-null values. In this case, if a row has 2 or fewer non-null values, it will be dropped from the dataframe. If a row has 3 or more non-null values, it will be kept in the dataframe. The inplace=False argument means that the original dataframe will not be modified, and a new dataframe with the dropped rows will be returned. If inplace=True, the original dataframe would be modified and the dropped rows would be removed permanently.
```

```python
df2.fillna(value={"Name": "", "Age": 0, "City": " ", "Salary": 000},inplace=False)
#fillna is used to fill the missing values in a dataframe. In this case, we are filling the missing values in the "Name" column with an empty string, the "Age" column with 0, the "City" column with a space, and the "Salary" column with 000. The inplace=False argument means that the original dataframe will not be modified, and a new dataframe with the filled values will be returned. If inplace=True, the original dataframe would be modified and the missing values would be filled permanently.
```

**merging joining and concatination**

```python
df5 = pd.DataFrame({
    "Emp_ID": [101, 102, 103, 104, 105],
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Eve"],
    "Department": ["HR", "IT", "Finance", "IT", "Marketing"]
},index=range(1,6))
df6 = pd.DataFrame({
    "Emp_ID": [101, 102, 103, 106],
    "Salary": [50000, 60000, 70000, 80000],
    "Bonus": [5000, 6000, 7000, 8000]
},index=range(1,5))
```

```python
pd.merge(df5, df6, on="Emp_ID", how="outer")
```

```python
pd.concat([df5, df6], axis=1)
```

```python
df6=pd.DataFrame({"name":["Rahul", "Priya"]},index=range(1,3))
df7=pd.DataFrame({"score":[85,90]},index=range(2,4))
```
